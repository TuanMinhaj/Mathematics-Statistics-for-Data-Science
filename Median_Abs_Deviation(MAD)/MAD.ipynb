Median Absolute Deviation(MAD)

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('movies.csv')
df.shape
```

Transformed data into million prefix

```python
#convert revenue as million

df['revenue']=df.revenue.apply(lambda x:x/1000000)
df
```

```python
#Assigning values

count,mean,std,*_=df.revenue.describe()
```

```python
count,mean,std
```

Outlier detection using Z score

```python
def get_z_score(value,mean,std):
    z_score=(value-mean)/std 
    return z_score
```

```python
#Apply Z score values for all columns 

df['Z_Score']=df.revenue.apply(lambda x:get_z_score(x,mean,std))
```

```python
df.head(2)
```

General guideline is to treat anything that has a z score of 3 or more as an outlier

```python
#Detecting outliers

df[df.Z_Score>3]
```

Outlier detection using modified Z score
So we got only 1 outlier using simple Z score. Based on the kind of analysis we are doing sometimes it might be better to use modified Z score as it will return more outliers.

```python
def get_mad(s):
    median = np.median(s)
    diff = abs(s-median)
    MAD = np.median(diff)
    return MAD
```

```python
MAD=get_mad(df.revenue)
median=np.median(df.revenue)
MAD,median
```

```python
def get_modified_z(x,MAD,median):
    modified_z=0.6745*(x-median)/MAD
    return modified_z
```

```python
df['Modified_Z']=df.revenue.apply(lambda x:get_modified_z(x,MAD,median))
```

```python
df.head(2)
```

General guideline for modified Z score is to use 3.5 as a thresold. i.e. anything that has a mod z score of 3.5 or more is an outlier

```python
df[df['Modified_Z']>3.5]
```
